```python
import xlwings as xw

# 請自行下載 Excel 並填入你的檔案路徑，例如：
# wb = xw.Book(r"C:\Users\你的名字\Downloads\sample.xlsx")
wb = xw.Book(r"請填入Excel完整路徑")
data_sheet = wb.sheets["raw"]
report_sheet = wb.sheets["report"]

# 讀取所有資料到 DataFrame
import pandas as pd
df = data_sheet.range("A1").options(pd.DataFrame, expand="table").value
#將同一名字的請假日排一起
df_sorted = df.sort_values(by=['員編', '請假日'])
df_sorted
df

# 出現次數
employee_counts = df.sort_values(by=['員編', '請假日']).value_counts()
employee_counts  

# 初始化儲存連續請假天數的字典
consecutive_days = {}

# 找出每位員工的連續請假天數
for name, group in df_sorted.groupby('姓名'):
    consecutive = 1
    max_consecutive = 1
    
    for i in range(1, len(group)):
        if (group.iloc[i]['請假日'] - group.iloc[i-1]['請假日']).days == 1:
            consecutive += 1
            max_consecutive = max(max_consecutive, consecutive)
        else:
            consecutive = 1
    
    if max_consecutive > 1:  
        consecutive_days[name] = max_consecutive

# 印出結果
print("連續請假天數:")
for name, days in consecutive_days.items():
    print(f"{name}\t{days}")
```
